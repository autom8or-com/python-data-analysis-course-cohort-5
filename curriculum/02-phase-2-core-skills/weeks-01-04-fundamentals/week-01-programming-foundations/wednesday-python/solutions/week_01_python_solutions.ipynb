## Exercise 1: Variables and Data Types Solutions

### Task 1.1: Create Customer Variables

```python
# Sample solution - students will have different values
customer_name = "Adebayo Okonkwo"       # String: Full name
customer_age = 32                       # Integer: Age in years  
customer_city = "Lagos"                 # String: Nigerian city
monthly_income = 350000.0               # Float: Income in Naira
has_smartphone = True                   # Boolean: True or False
years_online_shopping = 4               # Integer: Years of experience

# Display customer profile
print("=== CUSTOMER PROFILE ===")
print(f"Name: {customer_name}")
print(f"Age: {customer_age} years old")
print(f"Location: {customer_city}")
print(f"Monthly Income: â‚¦{monthly_income:,}")
print(f"Smartphone User: {has_smartphone}")
print(f"Online Shopping Experience: {years_online_shopping} years")
```

### Task 1.2: Business Calculations - Complete Solution

```python
# Calculate customer spending potential and category
# Using variables from Task 1.1

# Calculate estimated monthly e-commerce budget (5% of income)
estimated_monthly_budget = monthly_income * 0.05

# Calculate annual shopping potential
annual_potential = estimated_monthly_budget * 12

# Determine customer category based on income
if monthly_income >= 500000:
    customer_tier = "Premium"
elif monthly_income >= 200000:
    customer_tier = "Gold"
elif monthly_income >= 100000:
    customer_tier = "Silver"
else:
    customer_tier = "Bronze"

# Determine marketing priority
high_priority = has_smartphone and years_online_shopping >= 2 and monthly_income >= 150000

print("=== BUSINESS ANALYSIS ===")
print(f"Estimated Monthly Budget: â‚¦{estimated_monthly_budget:,.2f}")
print(f"Annual Spending Potential: â‚¦{annual_potential:,.2f}")
print(f"Customer Tier: {customer_tier}")
print(f"High Priority for Marketing: {high_priority}")

# Student task solution: Shopping frequency score calculation
# Age factor: 1.0 if age 25-45, 0.8 if younger/older
age_factor = 1.0 if 25 <= customer_age <= 45 else 0.8

# Experience: years_online_shopping / 5 (max 1.0)
experience_factor = min(years_online_shopping / 5, 1.0)

# Smartphone bonus: 1.2 if has smartphone, 1.0 if not
smartphone_bonus = 1.2 if has_smartphone else 1.0

# Final score
shopping_frequency_score = age_factor * experience_factor * smartphone_bonus

print(f"Shopping Frequency Score: {shopping_frequency_score:.2f}")
print(f"\n=== SCORE BREAKDOWN ===")
print(f"Age Factor: {age_factor} (ideal age range: 25-45)")
print(f"Experience Factor: {experience_factor:.2f} (max 1.0 for 5+ years)")
print(f"Smartphone Bonus: {smartphone_bonus} (1.2x if has smartphone)")
```

### Task 2.2: Sales Analysis - Complete Solution

```python
# Monthly sales data for 6 months
months = ["July", "August", "September", "October", "November", "December"]
sales_figures = [2500000, 2750000, 2200000, 3100000, 2900000, 3800000]

# Find business insights
print("=== SALES PERFORMANCE ANALYSIS ===")

# Basic statistics
total_sales = sum(sales_figures)
average_monthly = total_sales / len(sales_figures)
best_month_sales = max(sales_figures)
worst_month_sales = min(sales_figures)

print(f"Total Sales (6 months): â‚¦{total_sales:,}")
print(f"Average Monthly Sales: â‚¦{average_monthly:,.2f}")
print(f"Best Month Sales: â‚¦{best_month_sales:,}")
print(f"Worst Month Sales: â‚¦{worst_month_sales:,}")

# Find which months had best/worst performance
best_month_index = sales_figures.index(best_month_sales)
worst_month_index = sales_figures.index(worst_month_sales)

print(f"\nBest performing month: {months[best_month_index]}")
print(f"Worst performing month: {months[worst_month_index]}")

# Student task: Calculate month-over-month growth rates
print("\n=== GROWTH ANALYSIS ===")
print("Month-over-Month Growth:")

positive_growth_months = []
negative_growth_months = []

for i in range(1, len(sales_figures)):
    current_month = sales_figures[i]
    previous_month = sales_figures[i-1]
    growth_rate = ((current_month - previous_month) / previous_month) * 100
    
    print(f"   {months[i]}: {growth_rate:.1f}% vs {months[i-1]}")
    
    # Categorize growth
    if growth_rate > 0:
        positive_growth_months.append(months[i])
    else:
        negative_growth_months.append(months[i])

print(f"\nMonths with positive growth: {len(positive_growth_months)} ({', '.join(positive_growth_months)})")
print(f"Months with negative growth: {len(negative_growth_months)} ({', '.join(negative_growth_months)})")

# Additional insights
h2_sales = sum(sales_figures[3:])  # Oct, Nov, Dec (H2)
h1_sales = sum(sales_figures[:3])  # July, Aug, Sep (H1)
h2_vs_h1_growth = ((h2_sales - h1_sales) / h1_sales) * 100

print(f"\n=== HALF-YEAR COMPARISON ===")
print(f"H1 Sales (Jul-Sep): â‚¦{h1_sales:,}")
print(f"H2 Sales (Oct-Dec): â‚¦{h2_sales:,}")
print(f"H2 vs H1 Growth: {h2_vs_h1_growth:.1f}%")
```

### Task 3.2: Business Intelligence - Complete Solution

```python
# Analyze the customer database for business insights
print("=== CUSTOMER DATABASE ANALYSIS ===")

# Basic statistics
total_customers = len(customers)
total_orders = sum(customer['total_orders'] for customer in customers)
total_revenue = sum(customer['lifetime_value'] for customer in customers)
vip_customers = sum(1 for customer in customers if customer['is_vip'])

print(f"Total Customers: {total_customers}")
print(f"Total Orders: {total_orders}")
print(f"Total Revenue: â‚¦{total_revenue:,}")
print(f"VIP Customers: {vip_customers} ({(vip_customers/total_customers)*100:.1f}%)")
print(f"Average Revenue per Customer: â‚¦{total_revenue/total_customers:,.2f}")

# City analysis
cities = [customer['city'] for customer in customers]
unique_cities = list(set(cities))

print(f"\n=== GEOGRAPHIC DISTRIBUTION ===")
print(f"Cities represented: {len(unique_cities)}")
for city in unique_cities:
    city_customers = [c for c in customers if c['city'] == city]
    city_revenue = sum(c['lifetime_value'] for c in city_customers)
    city_orders = sum(c['total_orders'] for c in city_customers)
    print(f"   {city}: {len(city_customers)} customers, â‚¦{city_revenue:,} revenue, {city_orders} orders")

# Student task: Category preference analysis
print(f"\n=== CATEGORY PREFERENCES ===")
categories = [customer['preferred_category'] for customer in customers]
unique_categories = list(set(categories))

for category in unique_categories:
    category_customers = [c for c in customers if c['preferred_category'] == category]
    category_revenue = sum(c['lifetime_value'] for c in category_customers)
    avg_category_revenue = category_revenue / len(category_customers)
    print(f"   {category}: {len(category_customers)} customers, â‚¦{category_revenue:,} revenue, â‚¦{avg_category_revenue:,.2f} avg")

# Student task: Age group analysis
print(f"\n=== AGE GROUP ANALYSIS ===")
young = [c for c in customers if c['age'] <= 30]  # 18-30
middle = [c for c in customers if 31 <= c['age'] <= 45]  # 31-45
mature = [c for c in customers if c['age'] >= 46]  # 46+

print(f"Young (18-30): {len(young)} customers")
if young:
    young_revenue = sum(c['lifetime_value'] for c in young)
    print(f"   Revenue: â‚¦{young_revenue:,}, Avg: â‚¦{young_revenue/len(young):,.2f}")

print(f"Middle (31-45): {len(middle)} customers")
if middle:
    middle_revenue = sum(c['lifetime_value'] for c in middle)
    print(f"   Revenue: â‚¦{middle_revenue:,}, Avg: â‚¦{middle_revenue/len(middle):,.2f}")

print(f"Mature (46+): {len(mature)} customers")
if mature:
    mature_revenue = sum(c['lifetime_value'] for c in mature)
    print(f"   Revenue: â‚¦{mature_revenue:,}, Avg: â‚¦{mature_revenue/len(mature):,.2f}")

# Student task: Find top performers
print(f"\n=== TOP CUSTOMER ANALYSIS ===")

# Customer with highest lifetime value
top_value_customer = max(customers, key=lambda x: x['lifetime_value'])
print(f"ðŸ† Highest Value: {top_value_customer['name']} (â‚¦{top_value_customer['lifetime_value']:,})")

# Customer with most orders
most_orders_customer = max(customers, key=lambda x: x['total_orders'])
print(f"ðŸ“¦ Most Orders: {most_orders_customer['name']} ({most_orders_customer['total_orders']} orders)")

# Customer with highest average order value
customers_with_avg = [(c, c['lifetime_value']/c['total_orders']) for c in customers]
highest_avg_customer, highest_avg_value = max(customers_with_avg, key=lambda x: x[1])
print(f"ðŸ’° Highest Avg Order: {highest_avg_customer['name']} (â‚¦{highest_avg_value:,.2f} avg)")
```

## Exercise 4: Working with DataFrames Solutions

### Task 4.1: Create and Explore DataFrame - Solution

```python
import pandas as pd
import numpy as np

# Create a comprehensive business dataset
business_data = {
    'order_id': ['ORD-1001', 'ORD-1002', 'ORD-1003', 'ORD-1004', 'ORD-1005', 
                 'ORD-1006', 'ORD-1007', 'ORD-1008', 'ORD-1009', 'ORD-1010'],
    'customer_name': ['Adebayo Okonkwo', 'Fatima Abdullahi', 'Chinedu Okoro', 'Amina Hassan', 'Yemi Adeyemi',
                      'Kemi Olatunji', 'Ahmed Bello', 'Grace Eze', 'Ibrahim Musa', 'Tolu Bakare'],
    'product_category': ['Electronics', 'Fashion', 'Electronics', 'Home & Garden', 'Books',
                        'Fashion', 'Electronics', 'Home & Garden', 'Fashion', 'Books'],
    'order_value': [125000, 45000, 890000, 78000, 12000, 
                    67000, 340000, 156000, 23000, 8500],
    'city': ['Lagos', 'Abuja', 'Port Harcourt', 'Kano', 'Ibadan',
             'Lagos', 'Abuja', 'Kano', 'Lagos', 'Ibadan'],
    'order_date': ['2024-01-15', '2024-01-16', '2024-01-17', '2024-01-18', '2024-01-19',
                   '2024-01-20', '2024-01-21', '2024-01-22', '2024-01-23', '2024-01-24'],
    'payment_method': ['Card', 'Transfer', 'Card', 'Cash', 'Transfer',
                       'Card', 'Card', 'Cash', 'Transfer', 'Transfer'],
    'is_vip': [True, False, True, False, False, False, True, False, False, False]
}

# Create DataFrame
df = pd.DataFrame(business_data)

print("=== DATASET CREATED ===")
print(f"Shape: {df.shape}")
print("\nFirst 5 rows:")
print(df.head())

print("\n=== DATASET INFO ===")
df.info()

print("\n=== BASIC STATISTICS ===")
print(df.describe())
```

### Task 4.2: Data Analysis Questions - Complete Solutions

```python
print("=== BUSINESS ANALYSIS QUESTIONS ===")

# Question 1: Total revenue by city
print("\n1. REVENUE BY CITY:")
city_revenue = df.groupby('city')['order_value'].sum().sort_values(ascending=False)
for city, revenue in city_revenue.items():
    print(f"   {city}: â‚¦{revenue:,}")

# Question 2: Payment method popularity (student task)
print("\n2. PAYMENT METHOD POPULARITY:")
payment_counts = df['payment_method'].value_counts()
payment_revenue = df.groupby('payment_method')['order_value'].sum().sort_values(ascending=False)
print("By frequency:")
for method, count in payment_counts.items():
    print(f"   {method}: {count} orders")
print("By revenue:")
for method, revenue in payment_revenue.items():
    print(f"   {method}: â‚¦{revenue:,}")

# Question 3: Average order value by category (student task)
print("\n3. AVERAGE ORDER VALUE BY CATEGORY:")
category_stats = df.groupby('product_category')['order_value'].agg(['mean', 'count', 'sum'])
for category, stats in category_stats.iterrows():
    print(f"   {category}: â‚¦{stats['mean']:,.2f} avg, {stats['count']} orders, â‚¦{stats['sum']:,} total")

# Question 4: VIP vs regular customers (student task)
print("\n4. VIP VS REGULAR CUSTOMERS:")
vip_stats = df.groupby('is_vip').agg({
    'order_value': ['count', 'sum', 'mean']
})
vip_stats.columns = ['Orders', 'Revenue', 'Avg_Order']
for vip_status, stats in vip_stats.iterrows():
    customer_type = "VIP" if vip_status else "Regular"
    print(f"   {customer_type}: {stats['Orders']} orders, â‚¦{stats['Revenue']:,} revenue, â‚¦{stats['Avg_Order']:,.2f} avg")

# Question 5: Date range of orders (student task)
print("\n5. ORDER DATE RANGE:")
min_date = df['order_date'].min()
max_date = df['order_date'].max()
date_range = pd.to_datetime(max_date) - pd.to_datetime(min_date)
print(f"   From: {min_date} to {max_date}")
print(f"   Duration: {date_range.days} days")
print(f"   Daily orders: {len(df) / (date_range.days + 1):.1f} per day")

# Question 6: Top 3 highest value orders
print("\n6. TOP 3 HIGHEST VALUE ORDERS:")
top_orders = df.nlargest(3, 'order_value')[['customer_name', 'product_category', 'order_value', 'city']]
for idx, order in top_orders.iterrows():
    print(f"   {order['customer_name']}: {order['product_category']} - â‚¦{order['order_value']:,} ({order['city']})")

# Question 7: Electronics orders analysis
print("\n7. ELECTRONICS ORDERS ANALYSIS:")
electronics_df = df[df['product_category'] == 'Electronics']
print(f"Electronics orders: {len(electronics_df)}")
print(f"Electronics revenue: â‚¦{electronics_df['order_value'].sum():,}")
print(f"Average electronics order: â‚¦{electronics_df['order_value'].mean():,.2f}")
print(f"Electronics customers: {electronics_df['customer_name'].tolist()}")

# Question 8: Custom analysis example
print("\n8. CUSTOM ANALYSIS: Lagos VIP Customers")
lagos_vip = df[(df['city'] == 'Lagos') & (df['is_vip'] == True)]
if len(lagos_vip) > 0:
    print(f"Lagos VIP customers: {len(lagos_vip)}")
    print(f"Lagos VIP revenue: â‚¦{lagos_vip['order_value'].sum():,}")
    print("Lagos VIP customer details:")
    for _, customer in lagos_vip.iterrows():
        print(f"   {customer['customer_name']}: {customer['product_category']} - â‚¦{customer['order_value']:,}")
else:
    print("No Lagos VIP customers found")
```

### Task 4.3: Business Insights Report - Complete Solution

```python
# Create a comprehensive business report
print("=== COMPREHENSIVE BUSINESS REPORT ===")
print(f"Analysis Date: {pd.Timestamp.now().strftime('%Y-%m-%d')}")
print(f"Data Period: {df['order_date'].min()} to {df['order_date'].max()}")

# Executive Summary
total_revenue = df['order_value'].sum()
total_orders = len(df)
avg_order_value = df['order_value'].mean()
unique_customers = df['customer_name'].nunique()

print(f"\n=== EXECUTIVE SUMMARY ===")
print(f"ðŸ’° Total Revenue: â‚¦{total_revenue:,}")
print(f"ðŸ“¦ Total Orders: {total_orders}")
print(f"ðŸ“Š Average Order Value: â‚¦{avg_order_value:,.2f}")
print(f"ðŸ‘¥ Unique Customers: {unique_customers}")
print(f"ðŸ“ˆ Revenue per Customer: â‚¦{total_revenue/unique_customers:,.2f}")

# Category Performance
print(f"\n=== CATEGORY PERFORMANCE ===")
category_stats = df.groupby('product_category').agg({
    'order_value': ['count', 'sum', 'mean']
}).round(2)
category_stats.columns = ['Orders', 'Revenue', 'Avg_Order']
category_stats = category_stats.sort_values('Revenue', ascending=False)

for category, row in category_stats.iterrows():
    revenue_pct = (row['Revenue'] / total_revenue) * 100
    print(f"   {category}:")
    print(f"      Orders: {int(row['Orders'])} ({int(row['Orders'])/total_orders*100:.1f}%)")
    print(f"      Revenue: â‚¦{row['Revenue']:,.0f} ({revenue_pct:.1f}%)")
    print(f"      Avg Order: â‚¦{row['Avg_Order']:,.2f}")

# Student task: Geographic Analysis
print(f"\n=== GEOGRAPHIC ANALYSIS ===")
city_stats = df.groupby('city').agg({
    'order_value': ['count', 'sum', 'mean'],
    'customer_name': 'nunique'
}).round(2)
city_stats.columns = ['Orders', 'Revenue', 'Avg_Order', 'Unique_Customers']
city_stats = city_stats.sort_values('Revenue', ascending=False)

print(f"Market penetration across {len(city_stats)} cities:")
for city, row in city_stats.iterrows():
    revenue_pct = (row['Revenue'] / total_revenue) * 100
    print(f"   {city}:")
    print(f"      Revenue: â‚¦{row['Revenue']:,.0f} ({revenue_pct:.1f}% of total)")
    print(f"      Orders: {int(row['Orders'])} orders from {int(row['Unique_Customers'])} customers")
    print(f"      Avg Order: â‚¦{row['Avg_Order']:,.2f}")

# Student task: Customer Insights
print(f"\n=== CUSTOMER INSIGHTS ===")
vip_customers = df[df['is_vip'] == True]
regular_customers = df[df['is_vip'] == False]

vip_count = len(vip_customers)
vip_revenue = vip_customers['order_value'].sum()
vip_avg = vip_customers['order_value'].mean()

regular_count = len(regular_customers)
regular_revenue = regular_customers['order_value'].sum()
regular_avg = regular_customers['order_value'].mean()

print(f"Customer segmentation analysis:")
print(f"   VIP Customers: {vip_count} orders ({vip_count/total_orders*100:.1f}%)")
print(f"      Revenue: â‚¦{vip_revenue:,} ({vip_revenue/total_revenue*100:.1f}% of total)")
print(f"      Avg Order: â‚¦{vip_avg:,.2f}")
print(f"   Regular Customers: {regular_count} orders ({regular_count/total_orders*100:.1f}%)")
print(f"      Revenue: â‚¦{regular_revenue:,} ({regular_revenue/total_revenue*100:.1f}% of total)")
print(f"      Avg Order: â‚¦{regular_avg:,.2f}")
print(f"   VIP Premium: {vip_avg/regular_avg:.1f}x higher average order value")

# Payment method analysis
payment_stats = df.groupby('payment_method').agg({
    'order_value': ['count', 'sum', 'mean']
}).round(2)
payment_stats.columns = ['Orders', 'Revenue', 'Avg_Order']
payment_stats = payment_stats.sort_values('Revenue', ascending=False)

print(f"\n=== PAYMENT METHOD ANALYSIS ===")
for method, row in payment_stats.iterrows():
    print(f"   {method}: {int(row['Orders'])} orders, â‚¦{row['Revenue']:,.0f} revenue, â‚¦{row['Avg_Order']:,.2f} avg")

print(f"\n=== KEY PERFORMANCE INDICATORS ===")
print(f"ðŸ“Š Revenue Concentration: Top city ({city_stats.index[0]}) = {city_stats.iloc[0]['Revenue']/total_revenue*100:.1f}% of revenue")
print(f"ðŸŽ¯ Category Focus: Top category ({category_stats.index[0]}) = {category_stats.iloc[0]['Revenue']/total_revenue*100:.1f}% of revenue")
print(f"ðŸ‘‘ VIP Impact: {vip_count/total_orders*100:.1f}% of orders generate {vip_revenue/total_revenue*100:.1f}% of revenue")
print(f"ðŸ’³ Payment Preference: {payment_stats.index[0]} is most popular ({payment_stats.iloc[0]['Orders']/total_orders*100:.1f}% of orders)")

# Student recommendations section
print(f"\n=== BUSINESS RECOMMENDATIONS ===")
print("Based on the data analysis:")
print(f"1. Focus on {category_stats.index[0]} category - highest revenue generator")
print(f"2. Expand VIP program - VIPs spend {vip_avg/regular_avg:.1f}x more per order")
print(f"3. Strengthen {city_stats.index[0]} operations - {city_stats.iloc[0]['Revenue']/total_revenue*100:.1f}% of total revenue")
print(f"4. Promote {payment_stats.index[0]} payment method - already preferred by customers")
print(f"5. Target customer acquisition in underperforming cities")

print(f"\nðŸ“ˆ Report generated successfully! Data-driven insights for strategic decisions.")
```
